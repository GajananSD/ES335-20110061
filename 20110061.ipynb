```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
import mpmath
```

## Q9. ##

```python
# Define the domain
x_values = np.arange(0.5, 100.0, 0.5)
```

### 1) *y=x* ###

```python
# Calculate corresponding y values
y1_values = x_values
# Plot the function
plt.plot(x_values, y1_values, label='y = x')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
# Calculate corresponding y values
y2_values = x_values**2
# Plot the function
plt.plot(x_values, y2_values, label='y = x^2')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
# Calculate corresponding y values
y3_values = (x_values**3)/100
# Plot the function
plt.plot(x_values, y3_values, label='y = x^3/100')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^3/100')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
# Calculate corresponding y values
y4_values = np.sin(x_values)
# Plot the function
plt.plot(x_values, y4_values, label='y = sin(x)')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(x)')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
# Calculate corresponding y values
y5_values = np.sin(x_values)/(x_values)
# Plot the function
plt.plot(x_values, y5_values, label='y = sin(x)/x')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(x)/x')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```

```python
# Calculate corresponding y values
y6_values = np.log(x_values)
# Plot the function
plt.plot(x_values, y6_values, label='y = log(x)')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = log(x)')
# Add grid
plt.grid(True)
# Show legend
plt.legend()
# Display the plot
plt.show()
```
